#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python

from scipy.stats import zscore

ds2 = ds[ds['level5'].isin(["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"])]

# Normalización Z-score por población
ds2.loc[:, 'precio_norm'] = ds2.groupby('level5')['price'].transform(zscore)


# Colores y etiquetas
colors = {
    "Fuenlabrada": "blue",
    "Leganés": "green",
    "Getafe": "red",
    "Alcorcón": "purple"
}

# Graficar histogramas
plt.figure(figsize=(10, 6))
for city in colors:
    subset = ds2[ds2['level5'] == city]
    plt.hist(subset['precio_norm'], bins=30, alpha=0.6, label=city, color=colors[city], density=True)

plt.title("Distribución normalizada de precios por población")
plt.xlabel("Precio normalizado (Z-score)")
plt.ylabel("Densidad")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

En el histograma se muestran que la mayoría de las propiedades en cada población tienen precios cercanos o ligeramente inferiores a su media local (Z-score negativo). Esto sugiere que, dentro de cada municipio, el mercado inmobiliario tiende a concentrarse en rangos de precio relativamente homogéneos. Sin embargo, Getafe presenta una mayor dispersión hacia valores positivos, lo que indica presencia de viviendas significativamente más costosas que el promedio local

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps

# Mapa centrado en Madrid
map = Map(center = (40.35, -3.8), zoom = 11.5, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
from ipyleaflet import Map, basemaps, Circle, LayerGroup
import matplotlib.cm as cm
import matplotlib.colors as colors

datos = {
    "Getafe": {"lat": 40.3082, "lon": -3.7320, "poblacion": 123},
    "Leganés": {"lat": 40.3270, "lon": -3.7635, "poblacion": 1},
    "Fuenlabrada": {"lat": 40.2902, "lon": -3.8035, "poblacion": 400},
    "Alcorcón": {"lat": 40.3492, "lon": -3.8285, "poblacion": 173625}

}
# map centred in madrid
mapa = Map(center=(40.35, -3.8),
           zoom=11.5,
           basemap=basemaps.OpenStreetMap.Mapnik)


poblaciones = [info["poblacion"] for info in datos.values()]
norm = colors.Normalize(vmin=min(poblaciones), vmax=max(poblaciones))
cmap = cm.get_cmap("plasma")

#  (circles with colors according to population)
circulos = []

for ciudad, info in datos.items():
    color = colors.to_hex(cmap(norm(info["poblacion"])))
    circulo = Circle(
        location=(info["lat"], info["lon"]),
        radius=400,
        color=color,
        fill_color=color,
        fill_opacity=0.3,
        name=ciudad
    )
    circulos.append(circulo)

# add circles to map
grupo = LayerGroup(layers=circulos)
mapa.add_layer(grupo)
mapa
```
